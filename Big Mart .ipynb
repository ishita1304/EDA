# Feature engineering

```python
sales['Item_ID_cat']=sales.Item_Identifier.str[0:2]
```

```python
sales.Item_Fat_Content.replace(to_replace=['LF','low fat','reg'],
                              value=['Low Fat','Low Fat','Regular'],
                              inplace=True)
```

```python
sales.loc[sales.Item_ID_cat=='NC','Item_Fat_Content'] = 'Non Edible'
```

```python
pd.crosstab(sales.Item_ID_cat,sales.Item_Fat_Content)
```

```python
sales.Item_Type.unique()
```

```python
perish=['Dairy', 'Meat', 'Fruits and Vegetables','Breakfast','Breads', 'Starchy Foods', 'Seafood']
```

```python
def funct(x):
    if x in perish:
        return('perishable')
    else:
        return('non_perishable')
```

```python
sales['Item_cat']=sales.Item_Type.apply(funct)
```

```python
sales.Outlet_Identifier.unique()
```

```python
lowes=['OUT010', 'OUT019']
high=['OUT027', 'OUT013']
```

```python
def outlets(x):
    if x in lowes:
        return('Low_performers')
    elif x in high:
        return('High_performers')
    else:
        return('Avg_performers')
```

```python
sales['Outlet_cat']=sales.Outlet_Identifier.apply(outlets)
```

```python
sales['Vintage']=2013-sales.Outlet_Establishment_Year
```

# Scaling & Transformation

```python
sales.columns
```

```python
finaldata=sales.drop(['Item_Identifier','Item_Type', 'Outlet_Identifier',
       'Outlet_Establishment_Year'],axis=1)
```

```python
finaldata
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
finaldata['scaled_sales']=sc.fit_transform(pd.DataFrame(finaldata.Item_Outlet_Sales))
```

```python
sc=StandardScaler()
finaldata['scaled_mrp']=sc.fit_transform(pd.DataFrame(finaldata.Item_MRP))
```

```python
sc=StandardScaler()
finaldata['scaled_weight']=sc.fit_transform(pd.DataFrame(finaldata.Item_Weight))
```

```python
sc=StandardScaler()
finaldata['scaled_visibility']=sc.fit_transform(pd.DataFrame(finaldata.Item_Visibility))
```

```python
sc=StandardScaler()
finaldata['scaled_vintage']=sc.fit_transform(pd.DataFrame(finaldata.Vintage))
```

```python
newdata=finaldata.drop(['Item_Weight','Item_MRP','Item_Outlet_Sales','Vintage','Item_Visibility'],axis=1)
```

```python
newdata['Outlet_Size']=newdata.Outlet_Size.map({'Small':1,'Medium':2,'High':3})
```

```python
dummydata=pd.get_dummies(newdata, drop_first=True)
```

```python
dummydata
```

```python
newdata.describe(include='object')
```

# Multi collinearity

```python
df_target=dummydata.scaled_sales
```

```python
feature_df=dummydata.drop(['scaled_sales'],axis=1)
feature_df.head()
```

```python
df_numeric=feature_df.loc[:,['Outlet_Size','scaled_mrp','scaled_weight','scaled_visibility','scaled_vintage']]
```

```python
df_numeric.corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variance_inflation_factor(df_numeric.values,4)
```

# Build the model

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
# Add constant
x=sm.add_constant(feature_df)
x.head()
```

# Train the model

```python
model=sm.OLS(df_target,x).fit()
model.summary()
```

```python
model.pvalues[model.pvalues>0.05].index
```
